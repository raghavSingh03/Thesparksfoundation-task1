```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('http://bit.ly/w-data')
data.head(6)
```

```python
#to check if data contains any null value
data.isnull == True
```

#There is no null value in the dataset

```python
sns.set_style('darkgrid')
sns.scatterplot(y =data['Scores'],x=data['Hours'])
plt.xlabel('Hours',size=14)
plt.ylabel('Marks',size=14)
plt.show()
```

```python
sns.regplot(x=data['Hours'],y= data['Scores'])
plt.title('Regression Plot', size=20)
plt.xlabel('Hours',size=14)
plt.xlabel('Marks',size=14)
plt.show()
print(data.corr())
```

# Training the model

```python
x =data.iloc[:,:-1].values
y =data.iloc[:,1].values
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state =0)
```

```python
regression =LinearRegression()
regression.fit(x_train,y_train)
print("Our model has been Trained!!")
```

# Predicting the model

```python
y_pred = regression.predict(x_test)
prediction =pd.DataFrame({'Hours':[i[0]for i in x_test],'Expected Marks':[k for k in y_pred]})
prediction
```

comparing the actual marks with expected marks

```python
compared_marks =pd.DataFrame({'Marks':y_test,'Expected Marks':y_pred})
compared_marks
```

# Evaluation of the model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print('Mean absolute Error: ',mean_absolute_error(y_test,y_pred))
```

```python
print('Mean squared Error: ',mean_squared_error(y_test,y_pred))
```

As we can see that our model has a slight value of the error,so we can conclude that their might be chances of biased prediction but the value will be significantly low

So at last lets Predict the percentage of an student based on given number of study hours.
